
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# group by gender
purchase_group = purchase_data.groupby("Gender")
# find out average price
purchase_group.mean()
```

```python
# finde out total purchase value 
purchase_group.sum()
```

```python
gender_analysis = pd.merge(purchase_group.mean(), purchase_group.sum(), on="Gender")
gender_analysis
```

```python
# remove the columns we don't need
gender_analysis = gender_analysis[["Price_x", "Price_y"]]
gender_analysis
```

```python
# add purchase count and normalized totals to gender_analysis
gender_analysis["Purchase Count"] = [113, 652, 15]
gender_analysis["Normalized Totals"] = gender_analysis["Price_y"]/gender_analysis["Purchase Count"]
gender_analysis
```

```python
# format the table
gender_analysis = gender_analysis.rename(columns={"Price_x":"Average Purchase Price", "Price_y":"Total Purchase Value",
                                                })
gender_analysis["Average Purchase Price"]=gender_analysis["Average Purchase Price"].astype(float).map(
    "${:,.2f}".format)
gender_analysis["Total Purchase Value"]=gender_analysis["Total Purchase Value"].astype(float).map(
    "${:,.2f}".format)
gender_analysis["Normalized Totals"]=gender_analysis["Normalized Totals"].astype(float).map(
    "${:,.2f}".format)
gender_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
# Categorize the existing players using the age bins.
purchase_data["Age Demographics"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
purchase_data.head()
```

```python
# use groupby to count how many people in each age bin
age_demographics = purchase_data.groupby("Age Demographics")
age_demographics = age_demographics.count()
age_demographics = age_demographics[["Age"]]
age_demographics = age_demographics.rename(columns={"Age":"Total Count"})
age_demographics
```

```python
# add percentage of players
age_demographics["Percentage of Players"] = age_demographics["Total Count"]/780*100
age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].astype(float).map(
    "{:,.2f}%".format)
age_demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group by SN and sort in descending order
spenders = purchase_data.groupby("SN")
spenders = spenders.count()
spenders.head()
```

```python
# get Purchase Count
spenders["Purchase Count"] = spenders["Purchase ID"]
spenders = spenders[["Purchase Count"]]
spenders.head()
```

```python
# find average price
avg_price = purchase_data.groupby("SN").mean()
avg_price.head()
```

```python
# add average price to spenders
spenders["Average Purchase Price"] = avg_price["Price"]
spenders.head()
```

```python
# find total price
total_price = purchase_data.groupby("SN").sum()
total_price.head()
```

```python
# add total price to spenders
spenders["Total Purchase Value"] = total_price["Price"]
spenders.head()
```

```python
# format the spenders
spenders = spenders.sort_values(by = "Purchase Count", ascending=False)
spenders["Average Purchase Price"] = spenders["Average Purchase Price"].astype(float).map("${:,.2f}".format)
spenders["Total Purchase Value"] = spenders["Total Purchase Value"].astype(float).map("${:,.2f}".format)
spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
# purchase_data.head()
items = purchase_data[["Item ID", "Item Name", "Purchase ID","Price"]]
items.head()
```

```python
# group items by item ID and ID name
items_group = items.groupby(["Item ID", "Item Name"])
most_pop_items = items_group.count()
most_pop_items = most_pop_items[["Purchase ID"]]
most_pop_items.head()
```

```python
# find item price
items_price = items_group.mean()
most_pop_items["Item Price"] = items_price["Price"]
most_pop_items.head()
```

```python
# find total price
total_purchase = items_group.sum()
most_pop_items["Total Purchase Value"] = total_purchase["Price"]
most_pop_items.head()
```

```python
# formating the most_pop_items
most_pop_items = most_pop_items.rename(columns={"Purchase ID": "Purchase Count"})
#most_pop_items = most_pop_items["Item Price"].astype(float).map("${:,.2f}".format)
#most_pop_items = most_pop_items["Total Purchase Value"].astype(float).map("${:,.2f}".format)
most_pop_items = most_pop_items.sort_values(by = ["Purchase Count"], ascending = False)
most_pop_items.head()
```
